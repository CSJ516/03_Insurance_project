```python
import pandas as pd
from pyecharts import Bar, Line, Overlap, Grid, Tree

import warnings
warnings.filterwarnings(action='ignore')
```

```python
doc = pd.read_excel('메리츠화재_약관.xlsx')
doc1 = doc.rename(columns={'Unnamed: 0':'항목'}).reset_index(drop=True)
doc1
```

```python
for i in range(0, len(doc1['항목'])):
    if '(3년월지급형)' in doc1['항목'][i]:
        doc1['항목'][i] = doc1['항목'][i].replace('(3년월지급형)','(3년)')
```

```python
for i in range(0, len(doc1['총지급금액(최소)'])):
    doc1['총지급금액(최소)'][i] = str(doc1['총지급금액(최소)'][i])[:-4]
    doc1['총지급금액(최대)'][i] = str(doc1['총지급금액(최대)'][i])[:-4]
```

```python
doc1
```

```python
# 기본계약 (Tree)

data = [
    {"children": [
            {"children": [
                    {"children": [],"name": "첫 보험료 납입시점"},
                   ],"name": doc1.columns[2]},
            {"children": [
                    {"children": [],"name": "최초 "+ doc1['보장 횟수'][0]},
                    ],"name": doc1.columns[5]},
            {"children": [
                    {"children": [],"name": str(doc1['총지급금액(최소)'][0]) +"만원"},
                    ],"name": doc1.columns[7]}
                  ],"name": doc1['항목'][0]}
]

tree = Tree("기본계약", width=1000, height=500)
tree.add("", data)
tree
```

```python
# 선택계약 - 보장개시일 (Tree)

data = [
    {"children": [
            {"children": [
                    {"children": [],"name": doc1['항목'][0]},
                    {"children": [],"name": doc1['항목'][3]},
                    {"children": [],"name": doc1['항목'][4]},
                    {"children": [],"name": doc1['항목'][5]},
                    {"children": [],"name": doc1['항목'][6]},
                    {"children": [],"name": doc1['항목'][9]},
                   ],"name": "첫 보혐료 납부 직후 보장 항목"},
            {"children": [
                    {"children": [],"name": doc1['항목'][1]},
                    {"children": [],"name": doc1['항목'][2]},
                    {"children": [],"name": doc1['항목'][7]},
                    {"children": [],"name": doc1['항목'][8]},
                    {"children": [],"name": doc1['항목'][10]},
                    {"children": [],"name": doc1['항목'][11]},
                    ],"name": "계약일부터 1년 이후 보장 항목"}
                  ],"name": doc1.columns[2]}
]

tree = Tree("선택계약(보장개시일)", width=1600, height=1000)
tree.add("", data)
tree
```

```python
for i in range(0, len(doc1['보장 횟수'])):
    if '제한없음' in doc1['보장 횟수'][i]:
        doc1['보장 횟수'][i] = '3회'
```

```python
doc1
```

```python
# 선택계약 - 보장횟수 (Bar)

attr = doc1['항목']
v1 = [doc1['보장 횟수'][i][0] for i in range(0, len(doc1['보장 횟수']))]

bar = Bar("선택계약(보장횟수)", extra_html_text_label=["※ 보장횟수에 제한이 없는 경우에는 3회로 표시"],
          width=1000, height=500)
bar.add("보장횟수", attr, v1, xaxis_rotate=30, yaxis_formatter='회')
bar
```

```python
doc2 = doc1[doc1['총지급금액(최소)'] == doc1['총지급금액(최대)']][['항목','총지급금액(최소)']]
doc2 = doc2.reset_index(drop=True)
doc2
```

```python
# 선택계약 - 보장금액 (Bar)

attr = doc2['항목']
v1 = doc2['총지급금액(최소)']

bar = Bar("선택계약(총지급금액)", width=1000, height=500)
bar.add("총지급금액(단위 : 만 원)", attr, v1, 
        mark_line=["average"], mark_point=["max", "min"],
        xaxis_rotate=30, yaxis_formatter='만원')
bar
```

```python
doc3 = doc1[doc1['총지급금액(최소)'] != doc1['총지급금액(최대)']][['항목','총지급금액(최소)','총지급금액(최대)']]
doc3
```

```python
# 선택계약 - 보장금액2 (Bar)

attr = doc3['항목']
v1 = doc3['총지급금액(최소)']
v2 = doc3['총지급금액(최대)']

bar = Bar("선택계약(총지급금액)", width = 1000, height=500)
bar.add("총지급금액(최소)", attr, v1, yaxis_formatter='만원')
bar.add("총지급금액(최대)", attr, v2, yaxis_formatter='만원')
bar
```

```python
for i in range(0, len(doc1)):
    globals()['v{}'.format(i+1)] = [doc1['보험료'][i]]
```

```python
# 보험료 계산 (Bar)

attr = ['보험료(단위 : 원)']

bar = Bar(width=1000, height=500)
bar.add("중증(기본계약)", attr, v1, is_stack=True, is_convert=True, is_label_show=True)
bar.add("상해응급실내원", attr, v2, is_stack=True, is_convert=True, is_label_show=True)
bar.add("신골절치료비", attr, v3, is_stack=True, is_convert=True, is_label_show=True)
bar.add("신깁스치료비", attr, v4, is_stack=True, is_convert=True, is_label_show=True)
bar.add("경증이상", attr, v5, is_stack=True, is_convert=True, is_label_show=True)
bar.add("경증 생활자금", attr, v6, is_stack=True, is_convert=True, is_label_show=True)
bar.add("경증 생활자금(3년)", attr, v7, is_stack=True, is_convert=True, is_label_show=True)
bar.add("중증 생활자금", attr, v8, is_stack=True, is_convert=True, is_label_show=True)
bar.add("중증 생활자금(3년)", attr, v9, is_stack=True, is_convert=True, is_label_show=True)
bar.add("중등도이상", attr, v10, is_stack=True, is_convert=True, is_label_show=True)
bar.add("중등도 생활자금", attr, v11, is_stack=True, is_convert=True, is_label_show=True)
bar.add("중등도 생활자금(3년)", attr, v12, is_stack=True, is_convert=True, is_label_show=True)
bar
```

```python
# 선택계약 - 질병별 지급률 (Pie)

from pyecharts import Pie
from pyecharts import Style

pie = Pie('신깁스치료비 지급률', title_pos='center', width=1000, height=500)
style = Style()
pie_style = style.add(
    label_pos="center",
    is_label_show=True,
    label_text_color=None
)

pie.add("", ["손가락캐스트", ""], [5, 95], center=[30, 30], radius=[18, 24], **pie_style)
pie.add("", ["단상지", ""], [10, 90], center=[50, 30], radius=[18, 24], **pie_style)
pie.add("", ["장상지", ""], [20, 80], center=[70, 30], radius=[18, 24], **pie_style)
pie.add("", ["단하지 외 4개", ""], [30, 70], center=[30, 70], radius=[18, 24], **pie_style)
pie.add("", ["장하지 외 4개", ""], [40, 60], center=[50, 70], radius=[18, 24], **pie_style)
pie.add("", ["고수상", ""], [60, 40], center=[70, 70], radius=[18, 24], legend_top="center", **pie_style)
pie
```
